```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('Customer_Segmentation.csv')
```

```python
df.head()
```

```python
df = df.drop(['Customer Id','Address'], 1)
```

```python
df.describe()
```

```python
df = df[df["Defaulted"].notna()]
```

```python
df.describe()
```

```python
df.corr()
```

```python
plt.figure(figsize = (10,8))
sns.heatmap(df.corr(), 
            annot=True, 
            vmin=-1, 
            vmax=1, 
            linewidth=.5,
            cmap='coolwarm')
```

```python
plt.figure(figsize = (10,8))
sns.scatterplot(x=df["Age"],y=df["Income"])
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i,random_state=101)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize = (12,10))
plt.plot(range(1,11),wcss, linestyle = '--', marker = 'o', markerfacecolor='red')
plt.xlabel('# of Clusters')
plt.ylabel('WCSS')
plt.title('K-means');
```

```python
kmeans = KMeans(n_clusters = 4,random_state = 101)
kmeans.fit(scaled_data)
```

```python
df_kmeans = df.copy()
df_kmeans['Segment K-means'] = kmeans.labels_
```

```python
df_anal = df_kmeans.groupby(['Segment K-means']).mean()
df_anal 
```

```python
df_kmeans["Legend"] = df_kmeans['Segment K-means'].map({0:'Low Risk Borrowers',
                                                        1:'Heavy Borrowers',
                                                        2:'High Risk Borrowers',
                                                        3:'Established Average'})
```

```python
plt.figure(figsize = (10,8))
sns.scatterplot(x=df_kmeans["Age"],y=df_kmeans["Income"], hue=df_kmeans["Legend"])
```

```python
from sklearn.decomposition import PCA
pca = PCA()
```

```python
pca.fit(scaled_data)
```

```python
pca.explained_variance_ratio_
```

```python
plt.figure(figsize = (12,10))
plt.plot(range(1,9),pca.explained_variance_ratio_.cumsum(),marker = 'o',linestyle = '--', markerfacecolor='red')
plt.title('Variance Explained')
plt.xlabel('# of Components')
plt.ylabel('Cumulative Explained Variance')
```

```python
pca = PCA(n_components =4)
```

```python
pca.fit(scaled_data)
```

```python
pca.components_
```

```python
df_comp = pd.DataFrame(data = pca.components_,
                          columns = df.columns.values,
                          index = ['Component 1','Component 2','Component 3', 'Component 4'])
df_comp
```

```python
plt.figure(figsize = (10,8))
sns.heatmap(df_comp,
            annot=True,
            vmin=-1,
            vmax=1,
            linewidth=.5,
            cmap='coolwarm')
```

```python
pca.transform(scaled_data)
```

```python
pca_scores = pca.transform(scaled_data)
```

```python
wcss = []
for i in range(1,11):
    kmeans_pca = KMeans(n_clusters =i,init ='k-means++',random_state=42)
    kmeans_pca.fit(pca_scores)
    wcss.append(kmeans_pca.inertia_)
```

```python
plt.figure(figsize = (12,10))
plt.plot(range(1,11), wcss, linestyle = '--', marker = 'o', markerfacecolor='red')
plt.xlabel('# of Clusters')
plt.ylabel('WCSS')
plt.title('K-means with PCA')
plt.show()
```

```python
kmeans_pca = KMeans(n_clusters = 4,random_state=101)
```

```python
kmeans_pca.fit(pca_scores)
```

```python
df_pca_kmeans = pd.concat([df.reset_index(drop=True),pd.DataFrame(pca_scores)],axis = 1)
df_pca_kmeans.columns.values[-4:] = ['Component 1','Component 2','Component 3','Component 4']
df_pca_kmeans['Segment K-means PCA'] = kmeans_pca.labels_
```

```python
df_pca_kmeans_freq = df_pca_kmeans.groupby(['Segment K-means PCA']).mean()
df_pca_kmeans_freq
```

```python
df_pca_kmeans['Legend'] = df_pca_kmeans['Segment K-means PCA'].map({0:'Heavy Borrowers',
                                                                    1:'Low Risk Borrowers',
                                                                    2:'High Risk Borrowers',
                                                                    3:'Established Average'})
```

```python
plt.figure(figsize = (12,10))
sns.scatterplot(x = df_pca_kmeans['Component 2'], 
                y = df_pca_kmeans['Component 1'],
                hue = df_pca_kmeans['Legend'], 
                palette='Set1')
```
